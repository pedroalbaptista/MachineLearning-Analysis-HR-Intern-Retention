## 1. Contexto e Problema de Negócio
O presente projeto insere-se na área de **Gestão de Capital Humano**.  
O objetivo principal é **identificar os fatores que influenciam a retenção de estagiários (`Retained`)**, permitindo que o departamento de **Recursos Humanos** tome decisões proativas.

A retenção de talentos é crítica para:
- Maximizar o **retorno sobre o investimento em formação**
- Garantir a **continuidade do pipeline de talentos**

<br>
<hr>

## 2. Perguntas de Negócio

Para orientar a análise, o projeto responderá às seguintes questões:

- **Drivers de Retenção**  
  Qual o impacto do **salário**, da **avaliação do supervisor** e da **duração do estágio** na decisão de permanência?  
  Os “melhores” estagiários (top performers) são os que saem mais rapidamente?

- **Análise por Departamento**  
  Existem departamentos com **taxas de retenção** significativamente acima ou abaixo da média?

- **Perfil do Estagiário**  
  Existe correlação entre a **assiduidade** e a probabilidade de retenção?  
  Quais são as características mais comuns dos estagiários que ficam na empresa?

- **Interações relevantes**  
  Existem perfis específicos (ex.: top performer em determinados departamentos) com maior risco de saída?

<br>
<hr>

## 3. Objetivos de Data Mining e Estratégia de Modelação
O problema será abordado como uma **Classificação Binária**:

- `0` → Sai  
- `1` → Fica  

Nesta fase **não será selecionado um algoritmo final** (como *Random Forest* ou *KNN*).  
Será definida apenas a **estratégia de modelação**:

- Utilização de **modelos de aprendizagem supervisionada**
- Comparação e validação dos modelos na **Fase 4 – Modeling**

Além da modelação preditiva, foi realizada uma análise exploratória detalhada e engenharia de features para capturar padrões não lineares e interações complexas entre as variáveis.

## 4. Definição e Justificação de Métricas

A escolha das métricas foi orientada pelo **custo de erro para o negócio**.
No contexto de retenção de estagiários, **nem todos os erros têm o mesmo impacto financeiro ou operacional**: o erro mais caro é o **Falso Negativo** (prever que o estagiário fica quando ele sai).
  
Consequência:
- A empresa **perde o talento**
- Não teve oportunidade de **intervir preventivamente**

### Métrica Principal: Recall da Classe 0 (Sai)
- **Porquê?** Ajuda a **minimizar estas "fugas"**..
- **Meta definida:** ≥ 0.78 (garantir que pelo menos 78% das saídas reais sejam detetadas).

### Métrica de Equilíbrio: F2-Score (Classe 0)
- Dá peso duplo ao Recall em relação à Precision (β = 2).
- Usado como critério de desempate.

### Métrica de Qualidade Mínima: ROC AUC
- **Meta:** ≥ 0.60 (garante capacidade discriminatória aceitável).

**Porquê não priorizar Precision ou Accuracy?**
- Precision foca-se em evitar **Falsos Positivos** (prever que alguém vai sair quando, na verdade, fica), pelo que se considera secundária, neste contexto. **Análise de custo:** Um falso positivo implica apenas uma intervenção preventiva desnecessária — custo de tempo e recursos marginal, insignificante quando comparado à perda definitiva de um talento. Ou seja, o erro de **Precision** é **menos penalizador** para o negócio.

- Accuracy não foi utilizada como métrica de avaliação. Apesar do **leve desbalanceamento** da target (61% Retained vs 39% Sai), esta métrica seria pouco informativa: um modelo "burro" que previsse sempre "fica" alcançaria automaticamente 61% de Accuracy sem identificar qualquer risco de saída.

## 5. Critérios de Sucesso

### Técnico
- Recall (Classe 0) ≥ 0.78 **E** ROC AUC ≥ 0.60
- F2-Score (Classe 0) como critério de desempate
- Modelo estável validado por Cross-Validation (Stratified 5-Fold)

### Negócio
- Identificar os **principais fatores de risco** de saída
- Permitir ao RH reduzir a taxa de churn através de intervenções proativas

## 6. Impacto Esperado (após implementação do modelo)

### Redução projetada do turnover

O modelo permite identificar antecipadamente uma proporção significativa dos estagiários com maior probabilidade de saída, possibilitando a adoção de medidas preventivas e reduzindo a taxa de turnover.

### Eficiência de custos

A diminuição do turnover contribui para a redução dos custos associados à substituição de estagiários, nomeadamente:

- Processos recorrentes de recrutamento e seleção
- Tempo e recursos envolvidos no onboarding e integração
- Perda de produtividade durante o período em que a vaga permanece por preencher

### Suporte à decisão baseado em dados

A utilização do modelo promove uma transição para decisões mais informadas e sustentadas em evidência empírica, incluindo:

- Ajustes em políticas de compensação e benefícios para perfis com maior risco de saída
- Revisão de programas de estágio em departamentos com maior incidência de turnover
- Monitorização contínua de indicadores relevantes, como assiduidade e avaliação de desempenho, enquanto sinais precoces de possível saída
