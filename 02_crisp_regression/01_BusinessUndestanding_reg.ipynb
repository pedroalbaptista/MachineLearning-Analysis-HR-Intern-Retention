## 1. Contexto e Problema de Negócio
O objetivo é desenvolver um modelo capaz de estimar o salário ideal para um estagiário com base no seu perfil de entrada e desempenho, eliminando a subjetividade no processo de atribuição salarial.

Diferente do modelo anterior de classificação, aqui o foco é a equidade e consistência financeira. Queremos entender se o salário é ditado por variáveis objetivas (como mérito e área de estudo) ou se existe um elevado grau de aleatoriedade na empresa.

<br>
<hr>

## 2. Perguntas de Negócio
- **Recompensa de Desempenho:** A Supervisor_Evaluation tem um peso estatisticamente relevante na variação do salário, ou os salários são fixos à partida?

- **Assimetrias Estruturais:** Existem disparidades salariais sistemáticas entre departamentos (Department) que não sejam justificadas pelo perfil ou competência do estagiário?

- **Consistência:** Conseguimos prever o salário de um novo candidato apenas com os dados disponíveis no momento da contratação/avaliação?

<br>
<hr>

## 3. Definição do Problema de Machine Learning
- **Tipo de Problema:** Regressão Supervisionada.

- **Variável Alvo (Target):** Salary (Valor numérico contínuo).

- **Variáveis Preditoras (Features):** Idade, Major, Internship_Duration_Months, Department, Supervisor_Evaluation e Attendance_%.

- **Exclusão Estratégica:** A variável Retained será excluída das variáveis explicativas. Embora faça parte do histórico, o seu uso causaria Data Leakage para um cenário de previsão de novos salários, uma vez que a retenção é um resultado posterior à definição salarial.

<br>
<hr>

## 4. Critérios de Sucesso

### Técnico
- **Métricas:** Minimização do MAE (Erro Médio Absoluto) e RMSE.

- **Benchmarking:** O modelo deve apresentar um desempenho superior a uma baseline simples (média salarial).

- **Interpretabilidade:** Identificar claramente o peso (coeficientes) de cada variável no cálculo do salário.

### Negócio
- Criação de uma grelha salarial preditiva que ajude o RH a evitar propostas desajustadas (muito altas ou muito baixas) que possam causar insatisfação ou custos desnecessários.

<br>
<hr>

## 5. Riscos e Suposições
- **Suposição de Independência Temporal:** Assume-se que a decisão salarial é tomada com base no perfil e performance, e que o facto de o estagiário ter ficado ou saído posteriormente (Retained) é uma consequência e não uma causa do salário.

- **Risco de Baixa Correlação:** Como detetado na fase de Data Understanding feita anteriormente no problema de classificação, as variáveis atuais podem ter uma relação fraca com o salário, sugerindo que o critério atual da empresa pode ser pouco estruturado (ruído elevado).
    - O Risco da "Média": Como os dados têm ruído, o modelo pode tender a sugerir sempre valores próximos da média, falhando em identificar talentos "premium" que merecem o teto salarial (2000€).
